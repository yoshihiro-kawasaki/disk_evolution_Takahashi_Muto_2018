```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.constants as sc
import linecache

## 物理定数

pi           = np.pi                    # PI
k_b          = sc.k*1e7                 # Boltzmann constant in erg/K
m_p          = sc.proton_mass*1e3       # proton mass in g
Grav         = sc.G*1e3                 # gravitational constant in cm^3 g^-1 s^-2
AU           = sc.au*1e2                # astronomical unit in cm
year         = sc.Julian_year           # year in s
mu           = 2.34e0                    # mean molecular mass in proton masses
M_sun        = 1.9891e+33               # mass of the sun in g
R_sun        = 69550800000.0            # radius of the sun in cm
L_sun        = 3.828000000000000255e+33
sigmaSB      = 5.670500000000000003e-05
```

```python
rmin = 1.0e-1
rmax = 1.0e3
N = 121
rb = np.logspace(np.log10(rmin), np.log10(rmax), N) 
rc = np.sqrt(rb[1:]*rb[:-1])
```

```python
Sigma = np.zeros(len(rc))
# Sigma[1:-1] = 1e4 * rc[1:-1]**(-12/7)
Sigma = 1e4 * rc**(-12/7)
```

```python
plt.figure(figsize=(12, 10))
plt.plot(rc, Sigma, label=r"Sigma")
plt.xscale("log")
plt.yscale("log")
plt.xlabel(r"$r \ \mathrm{[au]}$", fontsize=30)
plt.ylabel(r"$\Sigma \ [\mathrm{g \ cm^{-2}}]$", fontsize=30)
plt.legend(fontsize=30)
plt.show()
```

```python
x  = rc[0]
x0 = rc[1]
x1 = rc[2]
x2 = rc[3]
y0 = Sigma[1]
y1 = Sigma[2]
y2 = Sigma[3]

m = (y1 - y0) / (x1 - x0)
Sigma0 = m*(x - x0) + y0
print(Sigma[0], Sigma0)
```

```python
x  = rc[0]
x0 = rc[1]
x1 = rc[2]
x2 = rc[3]
y0 = Sigma[1]
y1 = Sigma[2]
y2 = Sigma[3]
r0 = (x - x1) * (x - x2) / ((x0 - x1) * (x0 - x2))
r1 = (x - x2) * (x - x0) / ((x1 - x2) * (x1 - x0))
r2 = (x - x0) * (x - x1) / ((x2 - x0) * (x2 - x1))

Sigma0 = r0 * y0 + r1 * y1 + r2 * y2
print(Sigma[0], Sigma0)
```

```python
x  = rc[0]
x0 = rc[1]
x1 = rc[2]
x2 = rc[3]
y0 = Sigma[1]
y1 = Sigma[2]
y2 = Sigma[3]

m = (np.log(y1) - np.log(y0)) / (np.log(x1) - np.log(x0))
print(m)
Sigma0 = y0 * (x/x0)**m

print(Sigma[0], Sigma0)
```

```python
dlnHdlnr_m1 = 1.49 - 1.0 

Ms   = 0.1 * M_sun
Rs   = 2.0 * R_sun
Mdot = 2.0e-5 * M_sun / year

logLs = 0.20 + 1.74*np.log10(Ms/M_sun)
Ls = (10.0**logLs) * L_sun
Lacc = Grav * Ms * Mdot / (2.0 * Rs)

aspect = 1.0e-1

print(Ls/L_sun, Lacc/L_sun)
```

```python
# Qirr = (Ls + Lacc) * aspect * dlnHdlnr_m1 / (4.0*np.pi*rc**2.0)
# Tirr = (Qirr/sigmaSB)**(0.25)
Tirr = (Ls + Lacc) / (4.0*np.pi*sigmaSB*rc**2.0) * (2.0*Rs/(3.0*np.pi*rc) + 0.5*aspect*dlnHdlnr_m1)
Tirr = Tirr**0.25
print(Tirr)
```

```python
Tmin = 10.0
Tmax = 1.5e4
N = 150
T = np.logspace(np.log10(Tmin), np.log10(Tmax), N)

TMRI = 1.0e3
alpha_active = 1e-2
alpha_dead = 1e-4

alpha_MRI = 0.5*(alpha_active - alpha_dead) * (1.0 - np.tanh((TMRI - T)/200)) + alpha_dead

plt.figure(figsize=(12, 10))
plt.plot(T, alpha_MRI)
plt.xscale("log")
plt.yscale("log")
plt.show()
```

```python
T = np.logspace(np.log10(1e1), np.log10(1e4), 200)
# print(T)

kappaR = np.zeros_like(T)

for i in range(len(kappaR)):
    kappaR[i] = 2.25 * (1.0 - np.tanh((T[i]-1.5e3)/150)) * min(1.0, (T[i]/150)**2.0)
```

```python
plt.figure(figsize=(12, 10))
plt.plot(T, kappaR)
plt.xscale("log")
plt.yscale("log")
plt.ylim(1e-5)
plt.show()
```

```python
def Zhu2012(rho, T, amax=None):
    """Opacity tables for PPD's from Zhu et al. (2012, [1]).
    An updated version of the Bell & Lin (1994, [2]) opacities (assuming ISM 
    grains).
    args:
        rho : float or array of float
            Density, c.g.s
        T : float or array of float
            Temperature, K
        amax : float or array of float, ignored
            Maximum grain size, cm       
    References:
        [1] Zhu, Z., Hartmann, L., Nelson, R. P., et al. 2012, ApJ, 746, 110.
        [2] Bell, K. R. & Lin, D. N.C. 1994, ApJ, 427, 987.
    """

    pre = rho * T * 8.314472e7 / 2.4
    xlp = np.log10(pre)
    xlt = np.log10(T)

    # Electron scattering
    xlop = -0.48

    # Water ice
    idx = xlt < 2.23567 + 0.01899 * (xlp - 5.)
    xlop = np.where(idx, 1.5 * (xlt - 1.16331) - 0.736364, xlop)
    done = idx

    # Ice Evaporation
    idx = (~done) & (xlt < 2.30713 + 0.01899 * (xlp - 5.))
    xlop = np.where(idx, -3.53154212 * xlt + 8.767726 - (7.24786 - 8.767726) * (xlp - 5.) / 16., xlop)
    done |= idx

    # Metal grains
    idx = (~done) & (xlt < 2.79055)
    xlop = np.where(idx, 1.5 * (xlt - 2.30713) + 0.62, xlop)
    done |= idx

    # Graphite Corrosion
    idx = (~done) & (xlt < 2.96931)
    xlop = np.where(idx, -5.832 * xlt + 17.7, xlop)
    done |= idx

    # Grain opacity
    idx = (~done) & (xlt < 3.29105 + (3.29105 - 3.07651) * (xlp - 5.) / 8.)
    xlop = np.where(idx, 2.129 * xlt - 5.9398, xlop)
    done |= idx

    # Silicate Evaporation
    idx = (~done) & (xlt < 3.08 + 0.028084 * (xlp + 4))
    xlop = np.where(idx, 129.88071 - 42.98075 * xlt + (142.996475 - 129.88071) * 0.1 * (xlp + 4), xlop)
    done |= idx

    # Water vapour
    idx = (~done) & (xlt < 3.28 + xlp / 4. * 0.12)
    xlop = np.where(idx, -15.0125 + 4.0625 * xlt, xlop)
    done |= idx

    # More water vapour
    idx = (~done) & (xlt < 3.41 + 0.03328 * xlp / 4.)
    xlop = np.where(idx, 58.9294 - 18.4808 * xlt + (61.6346 - 58.9294) * xlp / 4., xlop)
    done |= idx

    # Molecular opacity
    idx = (~done) & (xlt < 3.76 + (xlp - 4) / 2. * 0.03)
    xlop = np.where(idx, -12.002 + 2.90477 * xlt + (xlp - 4) / 4. * (13.9953 - 12.002), xlop)
    done |= idx

    # H scattering
    idx = (~done) & (xlt < 4.07 + (xlp - 4) / 2. * 0.08)
    xlop = np.where(idx, -39.4077 + 10.1935 * xlt + (xlp - 4) / 2. * (40.1719 - 39.4077), xlop)
    done |= idx

    # Bound-free, free-free
    idx = (~done) & (xlt < 5.3715 + (xlp - 6) / 2. * 0.5594)
    xlop = np.where(idx, 17.5935 - 3.3647 * xlt + (xlp - 6) / 2. * (17.5935 - 15.7376), xlop)
    done |= idx

    # Extra limits
    idx = (xlop < 3.586 * xlt - 16.85) & (xlt < 4.)
    xlop = np.where(idx, 3.586 * xlt - 16.85, xlop)

    return 10. ** xlop
```

```python
rho = 1e-9

for i in range(len(T)):
    kappaR[i] = Zhu2012(rho, T[i])
        
plt.figure(figsize=(12, 10))
plt.plot(T, kappaR)
plt.xscale("log")
plt.yscale("log")
plt.ylim()
plt.show()
```
